```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_targets=1,n_informative=1,noise=80,random_state=13)
```

```python
x
```

```python
y
```

```python
li = LinearRegression()
li.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,li.predict(x),color='red')
```

```python
print("m =",li.coef_)
print("b =",li.intercept_)
```

```python
m = li.coef_
b = -100
lr = 0.01
epoch = 200
```

```python
for i in range(epoch):
    slope = -2*np.sum(y-m*x.ravel()-b)
    stepsize = lr*slope
    b = b-stepsize
    y_prd = m*x+b
    plt.plot(x,y_prd)
    
plt.scatter(x,y)
```
